## 3 Replacing the missing values(AGE) based on SEX AND Pclass.

```python
def new_in(cols):
    Age = cols[0]
    Sex = cols[1]
    Pclass = cols[2]
    
    if pd.isnull(Age) :
            
            if(Pclass==1): 
                if(Sex=='male'):
                    return 40
                else : 
                    return 36
                
            
            elif(Pclass==2): 
                if(Sex=='male'):
                    return 30
                else :
                    return 28          
            
            
            
            else: 
                return 23
            
       
    else : return Age
            
```

```python
train['Age'] = train[['Age','Sex','Pclass']].apply(new_in,axis=1)
```
